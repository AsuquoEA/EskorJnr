```python
#To find total goals scored in the season:

Total_goals = epl['Goals'].sum()
print(Total_goals)
```

```python
#Another way of calculating the total goals scored without manually typing print statement. This is shown below.....

Total_goals = epl['Goals'].sum()
Total_goals
```

```python
#To find total penalty goals scored in the season: 

Penalty_goals = epl['Penalties'].sum()
Penalty_goals 
```

```python
#Pie chart for goals scored vs penalty 

plt.figure(figsize=(10, 6))
data = [Total_goals, Penalty_goals]
labels = ['Goals', 'Penalties']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```
