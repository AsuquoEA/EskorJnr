```python
#Plotting feature distribution for bar, KDE and line
ax = election.plot(kind='bar', x = 'STATE', y = 'title='2015 Nigeria presidential election')

#To label axis x and y:
ax.set_xlabel('STATE')
ax.set_ylabel('number of votes')
```

```python
#To plot in histogram:
ax = election.plot(kind='hist', title='2015 Nigeria presidential election')

#To label axis for x and y
ax.set_xlabel('states')
ax.set_ylabel('number of votes')
```

```python
#To plot in kde:
ax = election.plot(kind='kde',title='2015 Nigeria presidential election ')

#To label axis for x and y
ax.set_xlabel('states')
ax.set_ylabel('number of votes')
```

```python

#To plot in line:
ax = election.plot(kind='line',title='2015 Nigeria presidential election ')

#To label axis for x and y
ax.set_xlabel('states')
ax.set_ylabel('number of votes')
```
