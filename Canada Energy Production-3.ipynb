## Essential import conventions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels as sm
import matplotlib.ticker as ticker
```

## Read .csv file 

```python
EM = pd.read_csv('Canada Energy Production.csv')
```

## Rename first column

```python
EM.rename(columns={"domestic energy production in Canada": "Energy Mix"}, inplace=True)
```

```python
EM.dtypes
```

## Plot data

```python
plt.figure(figsize=(12,7))
for mix in EM['Energy Mix'].unique():
    subset = EM[EM['Energy Mix'] == mix]
    plt.bar(subset['Year'], subset['Value'], label=mix)
plt.title('Domestic Energy Production in Canada by Energy Type')
plt.xlabel('Year')
plt.ylabel(EM['Units'].iloc[0])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Turn off scientific notation on the y-axis
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

## For coal

```python
# Filter for coal only
coal = EM[EM['Energy Mix'] == 'Coal and coal products']

plt.figure(figsize=(10,6))
plt.plot(coal['Year'], coal['Value'], marker='o', linewidth=2, color='brown')

plt.title('Domestic Energy Production in Canada â€“ Coal and Coal Products')
plt.xlabel('Year')
plt.ylabel(coal['Units'].iloc[0])
plt.grid(True, linestyle='--', alpha=0.7)

# Remove exponential notation
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```
